测试一个训练集

```python
import h5py
from ase import Atoms

# 打开HDF5文件
f = h5py.File('SPICE.hdf5', 'r')
frames = []
group = f['103147721']

# 读取原子的类型和位置
atomic_numbers = group['atomic_numbers'][:]
conformations = group['conformations'][:]
# 读取原子的力和体系的能量
forces = group['dft_total_gradient'][:]
energies = group['dft_total_energy'][:]

# 对于每个构象，创建一个Atoms对象
for conformation, force, energy in zip(conformations, forces, energies):
    atoms = Atoms(numbers=atomic_numbers, positions=conformation)
    atoms.info['forces'] = force
    atoms.info['energy'] = energy
    frames.append(atoms)

# 关闭文件
f.close()
```

打印测试训练集

```python
from wizard.frames import MultiMol
MultiMol(frames).dump_sequence('test.xyz')
```

转化全部训练集

```python
import h5py
from ase import Atoms
from wizard.frames import MultiMol

# 打开HDF5文件
f = h5py.File('SPICE.hdf5', 'r')
frames = []

for group_name in f.keys():
    group = f[group_name]

    # 读取原子的类型和位置
    atomic_numbers = group['atomic_numbers'][:]
    conformations = group['conformations'][:]
    # 读取原子的力和体系的能量
    forces = group['dft_total_gradient'][:]
    energies = group['dft_total_energy'][:]

    # 对于每个构象，创建一个Atoms对象
    for conformation, force, energy in zip(conformations, forces, energies):
        atoms = Atoms(numbers=atomic_numbers, positions=conformation)
        atoms.info['forces'] = force
        atoms.info['energy'] = energy
        frames.append(atoms)
# 关闭文件
f.close()
# 打印结构
MultiMol(frames).dump_sequence('train.xyz')
```
